# Simple Imputer

```python
from sklearn.impute import SimpleImputer
```

```python
X = np.array([[10, 3],
              [0, 4],
              [5, 3],
              [np.nan, 3]])
```

```python
imputer = SimpleImputer(missing_values=np.nan,
              strategy='mean')

imputer.fit_transform(X)
```
